```python
df = pd.read_excel('ded_df.xlsx', dtype = {'sleep_quality': str, 'stress_level': str}).iloc[:,1:]
```

```python
sbn.set_context("paper", rc={"axes.labelsize":16})
sbn.pairplot(df.loc[:,['age',
                     'sleep_duration',
                     'heart_rate',
                     'daily_steps',
                     'physical_activity',
                     'height',
                     'weight',
                     'average_screen_time',
                     'systolic_bp',
                     'diastolic_bp']], diag_kind="kde", corner = True)
plt.show()
```

```python
sbn.set_context("paper", rc={"axes.labelsize":16}, font_scale = 1.25)
ax = sbn.catplot(x="dry_eye_disease", kind="count", order = ['Y', 'N'], data=df)
ax.set(xlabel='Dry Eye Disease',
       xticklabels = [ 'Yes', 'No'],
       ylabel='Count')

plt.show()
#plt.savefig('viz_DYD_disease.png')
```

```python
catplot_Y_N = ["sleep_disorder", "wake_up_during_night", "sleepy_during_day",
               "caffeine_consumption", "alcohol_consumption", "smoking", "medical_issue", "ongoing_medication", 
               "device_before_bed", "blue-light_filter", "discomfort_eye-strain", "redness_in_eye", "irritation_in_eye"]
```

```python
catplot_Likert = ["sleep_quality", "stress_level"]
```

```python
catplot_F_M = ["gender"]
```

```python
# Plot gender
for cat in catplot_F_M:    
    sbn.catplot(x=cat, kind="count", hue="dry_eye_disease", order = ['F', 'M'], data=df)
```

```python
# Plot likert categories
for cat in catplot_Likert:    
    sbn.catplot(x=cat, kind="count", hue="dry_eye_disease", order = ['1','2','3','4','5'], data=df)
```

```python
# Plot yes/no categories
for cat in catplot_Y_N:    
    sbn.catplot(x=cat, kind="count", hue="dry_eye_disease", order = ['Y', 'N'], data=df)
```

```python
fig, ax = plt.subplots(figsize=(9, 9))
ax.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
sbn.heatmap(df.loc[:,['age',
         'sleep_duration',
         'heart_rate',
         'daily_steps',
         'physical_activity',
         'height',
         'weight',
         'average_screen_time',
         'systolic_bp',
         'diastolic_bp']].corr(), cmap = 'coolwarm', ax=ax, linecolor = 'black', linewidths = 0.1)
plt.savefig('viz_corr.png')
#plt.show()
```
