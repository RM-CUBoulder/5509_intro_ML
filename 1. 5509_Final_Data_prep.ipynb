```python
df = pd.read_csv('Dry_Eye_Dataset.csv', sep = ',')
# columns in lower case
df.columns = map(str.lower, df.columns)
# replace tab with '_'
df.rename(columns=lambda name: name.replace(' ', '_'), inplace = True)
# list of column names for reference
reference = list(df)
# shorthen the name of long column names
df.rename(columns={'itchiness/irritation_in_eye': 'irritation_in_eye', 
                   'feel_sleepy_during_day':'sleepy_during_day',
                   'smart_device_before_bed': 'device_before_bed'
                  }, inplace = True)
# split column blood_pressure into components systolic blood pressure and diastolic blood pressure 
df = df.join(df['blood_pressure'].str.split('/', n=1, expand=True).rename(columns={0:'systolic_bp', 1:'diastolic_bp'}))
```

```python
# remove blood_pressure from data frame
df = df.loc[:,['gender',
             'age',
             'sleep_duration',
             'sleep_quality',
             'stress_level',
             'heart_rate',
             'daily_steps',
             'physical_activity',
             'height',
             'weight',
             'sleep_disorder',
             'wake_up_during_night',
             'sleepy_during_day',
             'caffeine_consumption',
             'alcohol_consumption',
             'smoking',
             'medical_issue',
             'ongoing_medication',
             'device_before_bed',
             'average_screen_time',
             'blue-light_filter',
             'discomfort_eye-strain',
             'redness_in_eye',
             'irritation_in_eye',
             'dry_eye_disease',
             'systolic_bp',
             'diastolic_bp']]
```

```python
# change datatype of systolic and diastolic blood pressure from object to numeric 
df[['systolic_bp','diastolic_bp']] = df[['systolic_bp','diastolic_bp']].apply(pd.to_numeric)
# change datatype of systolic and diastolic blood pressure from numeric to object 
df[['sleep_quality', 'stress_level']] = df[['sleep_quality', 'stress_level']].astype(str)
```
